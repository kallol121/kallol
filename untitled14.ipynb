```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# 1. Load the CSV file
df = pd.read_csv("SET-12 (1).csv")

# 2. Fill missing values with mean of the column
df.fillna(df.mean(numeric_only=True), inplace=True)

# 3. Identify and drop duplicate records
df.drop_duplicates(inplace=True)

# 4. Normalize numerical columns (Min-Max Scaling)
numerical_cols = ['Math', 'Science', 'Age']
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# 5. Discretize 'Math' scores into 3 categories using binning
df['Math_binned'] = pd.cut(df['Math'], bins=3, labels=["Low", "Medium", "High"])

# 6. Remove noise using binning method (smoothing by bin means on 'Science')
science_sorted = df.sort_values(by='Science')
bins = pd.cut(science_sorted['Science'], bins=3)
science_smoothed = science_sorted.groupby(bins)['Science'].transform('mean')
df.loc[science_sorted.index, 'Science'] = science_smoothed

# Display final DataFrame
df
```
